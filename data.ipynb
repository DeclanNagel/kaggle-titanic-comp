```python
decks_if_dead = ["G", "A"]
decks_if_survived = ["C", "E", "D", "B", "F"]

def fill_deck(data):
    if data['Deck'] == "n":
        if data['Survived'] == 0:
            return np.random.choice(decks_if_dead)
        else:
            return np.random.choice(decks_if_survived)
    return data['Deck']

def drop_na(data):
    data.drop(columns=["Ticket", "Cabin", "Name", "Embarked"], inplace=True, errors="ignore")
    return data

def fill_age(data):
    data["Age"] = data["Age"].fillna(data["Age"].mean())
    return data

def sex_mapping(data):
    data['Sex'] = data['Sex'].map({'male': 0, 'female': 1})
    return data

def deck_encoding(data):
    data = pd.get_dummies(data, columns=['Deck'], dtype=int)
    return data

def preprocess_data(data):
    data = data.copy()
    # dropping columns
    data = drop_na(data)
    # fill in null values
    data = fill_age(data)
    #data['Deck'] = data.apply(fill_deck, axis=1)
    # map non numerical values
    data = sex_mapping(data)
    # one hot encoding Deck values
    #data = deck_encoding(data)
    data.drop("Deck", axis=1, inplace=True)
    data.drop("PassengerId", axis=1, inplace=True)

    return data

filt_train_df = preprocess_data(train_data)
```

```python
deck_choices = ['A', 'B', 'C', 'D', 'E', 'F', 'G']
def fill_deck_for_test(data):
    data['Deck'] = data['Deck'].apply(lambda x: np.random.choice(deck_choices) if pd.isna(x) or x == 'n' else x)
    return data

def drop_na_for_test(data):
    data.drop(columns=["Ticket", "Cabin", "Name", "Embarked"], inplace=True, errors="ignore")
    return data

def fill_age_for_test(data):
    data["Age"] = data["Age"].fillna(data["Age"].mean())
    return data

def sex_mapping_for_test(data):
    data['Sex'] = data['Sex'].map({'male': 0, 'female': 1})
    return data

def deck_encoding_for_test(data):
    data = pd.get_dummies(data, columns=['Deck'], dtype=int)
    return data

def pre_process_for_test(data):
    #data = fill_deck_for_test(data)
    data.drop("Deck", axis=1, inplace=True)
    data.drop("PassengerId", axis=1, inplace=True)
    data = drop_na_for_test(data)
    data = fill_age_for_test(data)
    data = sex_mapping_for_test(data)
    #data = deck_encoding_for_test(data)
    return data

filt_test_data = pre_process_for_test(test_data)
filt_test_data.head(10)
```
